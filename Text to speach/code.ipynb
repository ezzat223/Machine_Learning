# `gTTS : ` google Text To Speach.

```python
from gtts import gTTS 
from googletrans import Translator
from translate import Translator
import IPython.display as ipd
from pygame import mixer 
from tempfile import TemporaryFile
```

```python
tts = gTTS("Hello, i'm Shehab Hegazy, a 22 years old senior.")
tts.save('hello.mp3')
```

```python
ipd.Audio('hello.mp3')
```

```python
tts2 = gTTS("      ", lang='ar')
tts2.save('arabic.mp3')
ipd.Audio('arabic.mp3')
```

```python
mixer.init()
sf = TemporaryFile()
tts.write_to_fp(sf)
sf.seek(0)
mixer.music.load('arabic.mp3')
# mixer.music.play()
```

```python
translator = Translator(to_lang="ar")
translation = translator.translate("Hello, i'm Shehab Hegazy, a 22 years old senior.")
translation
```

```python
import win32com.client
speaker = win32com.client.Dispatch("SAPI.SpVoice")

while 1:
    print("Enter a word or sentence : ")
    s = input()
    speaker.Speak(s)
    if s=="exit": break
```
